```python
import pandas as pd
import requests
```

```python
url = "https://pt.wikipedia.org/wiki/Lista_de_filmes_de_maior_bilheteria"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.raise_for_status()
```

```python
tabelas = pd.read_html(response.text)

tabelas
```

```python
tabela = tabelas[0]

tabela
```

```python
tabela_filtrada = tabela[["Diretor(a)", "Bilheteria (US$)"]]

tabela_filtrada["Bilheteria (US$)"] = tabela_filtrada["Bilheteria (US$)"].str.replace(r"\D", "", regex=True).astype("int64") 

tabela_filtrada.groupby("Diretor(a)")["Bilheteria (US$)"].sum().sort_values(ascending=False)[:10]

```

```python
from matplotlib import pyplot as plt
from matplotlib import rc
```

```python
top10 = (tabela_filtrada.groupby("Diretor(a)")["Bilheteria (US$)"].sum().sort_values(ascending=False)[:10])

rc("font", size = 16)

fig, ax = plt.subplots(figsize=(29, 15))
ax: plt.Axes

ax.set_title("Top 10 Diretores")
ax.set_xlabel("Diretor(a)")
ax.set_ylabel("Bilheteria Total (US$)")

bars = ax.bar(top10.index, top10.values)
ax.bar_label(bars, fmt='${:,.0f}', size = 12)

ax.grid(axis = 'y')
ax.tick_params('x', labelrotation = 45)
ax.yaxis.set_major_formatter('${x:,.0f}')

plt.tight_layout()
plt.show()
```
